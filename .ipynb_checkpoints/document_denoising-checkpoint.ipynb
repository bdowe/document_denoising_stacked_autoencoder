```python
import os
import glob

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing import image
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import UpSampling2D
from tensorflow.keras.layers import Reshape

from sklearn.model_selection import train_test_split
```

```python
def im_show(im_name):
    plt.figure(figsize=(20,8))
    img = cv2.imread(im_name, 0)
    plt.imshow(img, cmap="gray")
    print(f"[INFO] Image shape: {img.shape} ")
im_show("datasets/train/77.png")
im_show("datasets/train_cleaned/77.png")
```

```python
TRAIN_IMAGES = glob.glob('datasets/train/*.png')
CLEAN_IMAGES = glob.glob('datasets/train_cleaned/*.png')
TEST_IMAGES = glob.glob('datasets/test/*.png')
print(f"[INFO] Number of train pictures: {len(TRAIN_IMAGES)}")
print(f"[INFO] Number of train_cleaned pictures: {len(CLEAN_IMAGES)}")
print(f"[INFO] Number of test pictures: {len(TEST_IMAGES)}")
```

```python
IMG_W = 258
IMG_H = 540
BS = 20
EPOCHS = 200
```

```python
from image_loader import ImageLoader
```

```python
x_train = ImageLoader.load_image(TRAIN_IMAGES, IMG_W, IMG_H)
y_train = ImageLoader.load_image(CLEAN_IMAGES, IMG_W, IMG_H)
x_test = ImageLoader.load_image(TEST_IMAGES, IMG_W, IMG_H)

print(f"[INFO] x_train shape: {x_train.shape}")
print(f"[INFO] y_train shape: {y_train.shape}")
print(f"[INFO] x_test shape: {x_test.shape}")
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.15, 
                                                  random_state=42)
print(f"[INFO] train shape: {x_train.shape} and {y_train.shape}")
print(f"[INFO] validation shape: {x_val.shape} and {y_val.shape}")
```

```python
from deep_conv_autoencoder import AE

autoencoder = AE.create_model(IMG_W, IMG_H)
autoencoder.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss',
                                       min_delta=0,
                                       patience=5,
                                       verbose=1, 
                                       mode='auto')
checkpoint = ModelCheckpoint('document_denoising.hdf5', monitor='val_loss', verbose=1, save_best_only=True, mode='min')
```

```python
history = autoencoder.fit(x_train, y_train,
                         batch_size=BS,
                         epochs=EPOCHS,
                         validation_data=([x_val, x_val], y_val),
                         callbacks=[early_stopping, checkpoint]
                        )
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
preds = autoencoder.predict(x_test)
```

```python
def plot_digits(*args):
    args = [x.squeeze() for x in args]
    n = min([x.shape[0] for x in args])
    
    plt.figure(figsize=(2*n, 2*len(args)))
    for j in range(n):
        for i in range(len(args)):
            ax = plt.subplot(len(args), n, i*n + j + 1)
            plt.imshow(args[i][j])
            plt.gray()
            ax.get_xaxis().set_visible(False)
            ax.get_yaxis().set_visible(False)

    plt.show()

plot_digits(x_test[:5], preds[:5])
```

```python
import os
import glob

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Model, load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing import image
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

from tensorflow.keras.layers import Input, Dense, Activation, BatchNormalization, Flatten, Conv2D, MaxPooling2D, Dropout, UpSampling2D, Reshape, concatenate

from sklearn.model_selection import train_test_split
```

```python
TRAIN_IMAGES = glob.glob('datasets/train/*.png')
CLEAN_IMAGES = glob.glob('datasets/train_cleaned/*.png')
TEST_IMAGES = glob.glob('datasets/test/*.png')
print(f"[INFO] Number of train pictures: {len(TRAIN_IMAGES)}")
print(f"[INFO] Number of train_cleaned pictures: {len(CLEAN_IMAGES)}")
print(f"[INFO] Number of test pictures: {len(TEST_IMAGES)}")
```

```python
IMG_W = 258
IMG_H = 540
BS = 20
EPOCHS = 1000
```

```python
def im_show(im_name, blur=False):
    plt.figure(figsize=(20,8))
    img = cv2.imread(im_name, 0)
    if blur:
        img = cv2.medianBlur(img, 3)
    plt.imshow(img, cmap="gray")
    print(f"[INFO] Image shape: {img.shape} ")
im_show("datasets/train/77.png")
im_show("datasets/train/77.png", True)
im_show("datasets/train_cleaned/77.png")
```

```python
autoencoder = load_model('document_denoising_1.hdf5')
```

```python
# Load training data using default images, opencv processing methods, and autoencoder
x_train_default = ImageLoader.load_image(TRAIN_IMAGES, IMG_W, IMG_H)
x_train_blur = ImageLoader.load_image_blur(TRAIN_IMAGES, IMG_W, IMG_H)
x_train_canny_dilate_erode = ImageLoader.load_image_canny_dilate_erode(TRAIN_IMAGES, IMG_W, IMG_H)
x_train_threshold_binary = ImageLoader.load_image_threshold(TRAIN_IMAGES, IMG_W, IMG_H, 'binary')
x_train_threshold_gaussian = ImageLoader.load_image_threshold(TRAIN_IMAGES, IMG_W, IMG_H)
x_train_threshold_mean = ImageLoader.load_image_threshold(TRAIN_IMAGES, IMG_W, IMG_H, 'adaptive_mean')
x_train_autoencoder = autoencoder.predict(x_train_default)

# Load test data using default images, opencv processing methods, and autoencoder
x_test_default = ImageLoader.load_image(TEST_IMAGES, IMG_W, IMG_H)
x_test_blur = ImageLoader.load_image_blur(TEST_IMAGES, IMG_W, IMG_H)
x_test_canny_dilate_erode = ImageLoader.load_image_canny_dilate_erode(TEST_IMAGES, IMG_W, IMG_H)
x_test_threshold_binary = ImageLoader.load_image_threshold(TEST_IMAGES, IMG_W, IMG_H, 'binary')
x_test_threshold_gaussian = ImageLoader.load_image_threshold(TEST_IMAGES, IMG_W, IMG_H)
x_test_threshold_mean = ImageLoader.load_image_threshold(TEST_IMAGES, IMG_W, IMG_H, 'adaptive_mean')
x_test_autoencoder = autoencoder.predict(x_test_default)

# Concatenate training and test data into single lists
x_train = x_train_default + x_train_blur + x_train_canny_dilate_erode + x_train_threshold_binary + x_train_threshold_gaussian + x_train_threshold_mean + x_train_autoencoder
x_test = x_test_default + x_test_blur + x_test_canny_dilate_erode + x_test_threshold_binary + x_test_threshold_gaussian + x_test_threshold_mean + x_test_autoencoder

# Load cleaned images as training labels
y_train = ImageLoader.load_image(CLEAN_IMAGES, IMG_W, IMG_H)
```

```python
print(x_train.shape)
```

```python
#### Need to make sure all X's and y's match up - do some tests
## https://stackoverflow.com/questions/49829023/train-test-split-with-multiple-features
```

```python
#Split training data into training and validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.15, random_state=42)
print(f"[INFO] train shape: {x_train.shape} and {y_train.shape}")
print(f"[INFO] validation shape: {x_val.shape} and {y_val.shape}")
```

```python
stacker = AE.create_model(IMG_W, IMG_H)
stacker.compile(optimizer='adam', loss='mse')
stacker.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss',
                                       min_delta=0,
                                       patience=5,
                                       verbose=1, 
                                       mode='auto')
checkpoint = ModelCheckpoint('stacker_denoising.hdf5', monitor='val_loss', verbose=1, save_best_only=True, mode='min')
```

```python
history = autoencoder.fit(x_train, y_train,
                         batch_size=BS,
                         epochs=EPOCHS,
                         validation_data=(x_val, y_val),
                         callbacks=[early_stopping, checkpoint]
                        )
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
stacker = load_model('stacker_denoising.hdf5')
decoded_imgs = stacker.predict(x_test)

n = 10
plt.figure(figsize=(20, 4))
for i in range(1, n+1):
#     # display original
#     ax = plt.subplot(2, n, i)
#     plt.imshow(x_test[i].reshape(28, 28))
#     plt.gray()
#     ax.get_xaxis().set_visible(False)
#     ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, i + n)
    plt.imshow(decoded_imgs[i].reshape(IMG_W, IMG_H))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
